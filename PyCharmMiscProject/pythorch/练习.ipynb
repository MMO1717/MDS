```python
import os
from torch.utils.data import Dataset
from PIL import Image

class MyDataset(Dataset):
    def __init__(self, root_dir, label_dir):
        self.root_dir = root_dir
        self.label_dir = label_dir
        self.path = os.path.join(self.root_dir, self.label_dir)

        # 获取所有文件列表
        all_files = os.listdir(self.path)

        # 过滤掉非图片文件（如 .DS_Store）
        # 这里的判断条件可以更严格，比如只保留 .jpg, .png 等后缀
        self.img_path = [file for file in all_files if not file.startswith('.')]

    def __getitem__(self, index):
        img_name = self.img_path[index]
        img_item_path = os.path.join(self.path, img_name)
        img = Image.open(img_item_path)
        label = self.label_dir
        return img, label

    def __len__(self):
        return len(self.img_path)

# 主函数保持不变
def main():
    # 假设您的数据集根目录和标签目录如下
    root_directory = '/Users/mm/PyCharmMiscProject/Dataset'
    label_directory = 'ants'

    # 检查路径是否存在，避免运行时错误
    full_path = os.path.join(root_directory, label_directory)
    if not os.path.exists(full_path):
        print(f"错误：指定路径 '{full_path}' 不存在。请检查您的数据集路径是否正确。")
        return

    # 创建自定义数据集的实例
    my_dataset = MyDataset(root_dir=root_directory, label_dir=label_directory)

    # 1. 打印数据集的大小
    print(f"数据集大小（样本总数）为: {len(my_dataset)}")

    # 2. 遍历数据集中的所有样本
    print("\n遍历并打印数据集中的每个样本：")
    for i in range(len(my_dataset)):
        img_data, img_label = my_dataset[i]

        print(f"样本 {i+1}:")
        print(f"  文件名: {my_dataset.img_path[i]}")
        print(f"  图片大小: {img_data.size}")
        print(f"  图片格式: {img_data.format}")
        print(f"  标签: {img_label}")
        print("-" * 20)

if __name__ == "__main__":
    main()
```
