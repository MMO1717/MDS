```python
import os  # 导入os模块，用于处理文件和目录

from torch.utils.data import Dataset  # 从torch.utils.data导入Dataset类，用于自定义数据集
from PIL import Image  # 从PIL库导入Image模块，用于图片读取和处理

class MyDataset(Dataset):  # 定义一个继承自Dataset的自定义数据集类
    def __init__(self, root_dir, label_dir):  # 初始化方法，接收根目录和标签目录
        self.root_dir = root_dir  # 保存根目录路径
        self.label_dir = label_dir  # 保存标签目录名称
        self.path = os.path.join(self.root_dir, self.label_dir)  # 拼接得到图片文件夹的完整路径
        self.img_path = os.listdir(self.path)  # 列出该目录下所有图片文件名，存入列表



    def __getitem__(self, index):  # 获取指定索引的数据
        img_name = self.img_path[index]  # 根据索引获取图片文件名
        img_item_path = os.path.join(self.path, img_name)  # 拼接得到图片的完整路径
        img = Image.open(img_item_path)  # 打开图片文件
        label = self.label_dir  # 设置标签为标签目录名
        return img, label  # 返回图片和标签

    def __len__(self):  # 返回数据集的大小
        return len(self.img_path)  # 返回图片文件的数量

ants_dataset = MyDataset(root_dir='/Users/mm/PyCharmMiscProject/Dataset', label_dir='ants')  # 创建MyDataset对象，指定根目录和标签目录
```
