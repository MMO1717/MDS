```python
import  torch
from d2l import torch as d2l
"导入d2l.synthetic_data()"
from torch.utils import data
"导入TensorDataset DataLoader Dataset"



true_w = torch.tensor([2, -3.4])
true_b = 4.2
features,labels=d2l.synthetic_data(true_w,true_b,1000)
"featrues表示的是1000个随机样本，labels表示的是真实的输出"
```

```python
def load_array(data_arrays,batch_size,is_train=True):
    dataset=data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset,batch_size,shuffle=is_train)
batch_size=10
data_iter=load_array((features,labels),batch_size)
```

```python
from  torch import nn
net=nn.Sequential(nn.Linear(2,1))

nn.init.normal_(net[0].weight, mean=0, std=0.01)
nn.init.zeros_(net[0].bias)
```

```python
loss=nn.MSELoss()
trainer=torch.optim.SGD(net.parameters(),lr=0.04)
"torch.optim 是 PyTorch 的优化器模块，里面有各种常见的优化算法"
"net.parameters() 会返回一个 可迭代对象，里面存着所有需要优化的参数。"
```

```python
num_epochs=3
for epoch in range(num_epochs):
    for x,y in data_iter:
        l=loss(net(x),y)
        "计算损失 L(w, b)"
        trainer.zero_grad()
        "清空旧梯度："
        l.backward()
        "计算新梯度"
        trainer.step()
        "更新参数"
    l=loss(net(features),labels)
    print(f'epoch {epoch + 1}, loss {l:f}')
```
